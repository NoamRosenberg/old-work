"The number of respondents who were not asked a question because it did not apply to them--that is, "valid skips (-4)"--is also depicted."
https://www.nlsinfo.org/content/cohorts/nlsy97/using-and-understanding-the-data/nlsy97-documentation

I take this to mean that -4 values were out of the job market at the time.

```python
ExpList = ExpList.replace(-4, value = 0)
print(ExpList.shape)
```

```python


def nonNegativeMean(row):
    rowsum = 0.0
    colcount = 0
    for column in ExpList.columns:
        while colcount < 15:
            if row[column] >= 0:
                rowsum += row[column]
                colcount += 1
        else:
            if row[column] >= 0:
                rowsum += row[column] 
                colcount += 2
            
    return (rowsum / colcount)
        

MeanExpPerYear = ExpList.apply(nonNegativeMean, axis = 1)
TotalExp = MeanExpPerYear * (2012 - 1979)
dietnlsy['Experience (weeks)'] = TotalExp
```

Diabetes can not be used as an explanatory variable in our model

```python
from statsmodels.sandbox.regression.gmm import IV2SLS

dis = dietdummy.drop(['Income', 'Reads Ingredi','Date of Interview', 
                                                           'Hispanic', 'Blond', 'Light eyes', 'Times ate Fast Food in past Week'], axis = 1)
resultIV = IV2SLS(dis['Log Income'], dis.drop(['Log Income', 'Diabetes'], axis = 1), dis.drop(['Log Income', 'Reads Nutri'], axis = 1)).fit()
resultIV
```

```python
resultIV.spec_hausman()
```

There is no evidence to suggest that Dietary Behaviours are endogenous to Income. Or In other words there is no reason to believe that Income has an effect on Dietary Behaviours.
